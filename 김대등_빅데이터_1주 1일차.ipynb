```python
import pandas as pd
import numpy as np
from numpy.random import randn

# 연습데이터 생성
np.random.seed(1234)
df= pd.DataFrame(randn(10,2), columns=['A','B'])
df

df['A'] # 열 선택 Series 반환 
df[['A']] #열 선택 DataFrame 반환
df[0:5] # 행 선택 

# loc를 이용한 데이터 선택
df.loc[0:7,['A']]
df.loc[:,['A']]
df.loc[3:8]
df.loc[3:3]

# iloc를 이용한 데이터 선택
df.iloc[0:7,[0]]
df.iloc[:,[0]]
df.iloc[3:8]
df.iloc[3:4]

# query를 이용한 데이터 선택
df.query('A>0.5')

# query와 동일한 slice 표현
df.query('A>B')
df[df.A>df.B] 

# query를 이용한 데이터 선택-복수조건
df.query('A>0.5 & B<0.0' )
df.query('A>0.5 | B<0.0' )
```

# 데이터 결합
## ex) 제철소 제강부에서 데이터를 받으면 월별로 나눠서 받을 수 있다. 각 데이터를 결합해야하는 경우가 발생할 수 있다.
## 그럴때는 데이터 결합 concatenate(연결), append(추가) , merge, join 4가지가 있다.

### concatenate
#### 위+ 아래 결합 axis=0, 좌+우 axis=1 : 칼럼명이 다를 경우 사용한다.(인덱스나 칼럼명이 중복될 수 있어서 상황에 따라reset 시켜줘야한다)
#### append 는 위 + 아래 결합만 가능하다
 
 






### concatenate
#### 위+ 아래 결합 axis=0, 좌+우 axis=1 : 칼럼명이 다를 경우 사용한다.(인덱스나 칼럼명이 중복될 수 있어서 상황에 따라reset 시켜줘야한다)
#### append 는 위 + 아래 결합만 가능하다
 

### merge (column- column, column-index, index-index) 
#### 결함방법 : how=inner (중복된 칼럼만 들어옴 / left / right (오른쪽꺼는 다가져오고 ) / outer 
#### 비교되는 대상의 차이가 있다면 inner빼고는 전부 결측치가 발생한다. (시험) 
#### 비교되는 대상에 차이가 없다면 전부 결측치가 발생하지 않는다. 


#### join (index - index, column-index)(텍스트분석 때 사용 명사를 뽑아낸다음 명사간에 space 주고 싶을 때 join으로 스페이스 줘서 사용한다) 


### 결론적으로 데이터를 결합할 때는  merge가 가장 유용하고 많이 사용된다. 

#### Tips. 통계를 쓰는지 안쓰는지 알 수 있는 방법 : 자유도(분모 : n-k-1) , p-value값이 있으면 통계 사용 , 머신러닝에서는 회귀분석에서 사용된다.(절편, 회귀계수 2개)
#### 자유도란 통계적 추정을할 때 표본자료 중 모집단에 대한 정보를 주는 독립적인 자료의 수를 말한다.
