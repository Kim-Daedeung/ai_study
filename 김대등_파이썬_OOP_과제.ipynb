```python
# 과제 1 : Set 만들기 

class SET : 
    def __init__(self, _list=[]): 
        self.list=[]
        for i in _list : 
            if not i  in self.list :
                self.list.append(i)


    def ADD(self,elem):
        if not elem in self.list :
            self.list.append(elem)
            
    def DISCARD (self, elem):
        if elem in self.list :
            self.list.remove(elem)
            
    def CLEAR(self):
        self.list=[]
        
    def __len__(self):
        return len(self.list)


    def __str__(self):
        return "{" + str(self.list)[1:-1] + "}" 
    

    def __contains__(self, elem) :
        if elem in self.list :
            return True
        else : 
            return False
        
    def __le__(self, other) :
        for i in self.list:
            if i not in other.list:
                return False
        return True        

    def __ge__(self, other) :
        for i in other.list:
            if i not in self.list:
                return False
        return True

    def __or__(self,other) :
        new_or=self.list[:]
        for i in other.list:
            if i not in self.list:
                new_or.append(i)
        return SET(new_or)

    def __and__(self,other):
        new=[]
        for i in other.list :
            if i in self.list :
                new.append(i)
        return SET(new)
    
    def __sub__(self,other):
        new2=[]
        for i in self.list :
            if i not in other.list:
                new2.append(i)
        return SET(new2)
    
    
    
    def __ior__(self, other):
        for i in other.list :
            if not i in self.list :
                self.list.append(i)
        return self
    
    
    def __iand__ (self, other) :
        new_iand=[]
        for i in other.list :
            if i in self.list :
                new_iand.append(i)
        self.list=new_iand
        return self
    
    
    def __isub__(self, other):
        new_isub = []
        for i in self.list:
            if i not in other.list:
                new_isub.append(i)
        self.list=new_isub
        return self
    

```

```python
a = SET([1, 2, 3])
b = SET([2, 3, 4])
print(a|b)
print(a)
print(b)
print(a&b)
print(a)
print(b)
print(a-b)


a = SET([1,2,3])
b = SET([3,4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()
```
